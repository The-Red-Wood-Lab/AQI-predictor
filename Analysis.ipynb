<h1>Model Suggestions for Nonlinear and Seasonal Data</h1>
<p>We have realized the following about the data:</p>
<ul>
    <li><strong>First:</strong> The data is nonlinear and has somewhat seasonal trends.</li>
    <li><strong>Second:</strong> We absolutely cannot use other values like NO to predict NO2 because most of the missing values are overlapping. If NO is missing, most probably NO2 is missing as well.</li>
</ul>

<p>To address these challenges, here are some model suggestions:</p>

<h2>Time Series Regression</h2>
<p>Specifically designed for time series data, these models incorporate lagged values of the target variable and other predictors to account for autocorrelation and seasonality. Examples include:</p>
<ul>
    <li><strong>ARIMA (AutoRegressive Integrated Moving Average)</strong></li>
    <li><strong>SARIMA (Seasonal ARIMA)</strong></li>
</ul>

<h2>Generalized Additive Models (GAM)</h2>
<p>Provide a flexible framework for modeling non-linear relationships between CO concentration and date. GAMs can incorporate smooth functions to capture complex patterns.</p>

```python
df_kolkata = df[df["City"] == "Kolkata"]

# Convert 'Date' column to datetime
df_kolkata['Date'] = pd.to_datetime(df_kolkata['Date'], errors='coerce')

# Plot the data
plt.figure(figsize=(20,5))
plt.title("Kolkata NO2")

# Plot scatter plots for different pollutants
sns.scatterplot(x="Date", y="PM10", data=df_kolkata, palette="Blues", label="PM10")
sns.scatterplot(x="Date", y="NO", data=df_kolkata, palette="Reds", label="NO")
sns.scatterplot(x="Date", y="NO2", data=df_kolkata, palette="Greens", label="NO2")

# Format the x-axis to display dates properly
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))  # '%b %Y' will display as 'Jan 2015'

plt.xticks(rotation=45)  # Rotate the x-axis labels for readability

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
